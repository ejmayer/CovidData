```python
import os
# import urllib2
import pandas as pd
import urllib.request, json
from pprint import pprint
from pandas.io.json import json_normalize
```

```python
with urllib.request.urlopen("https://corona-api.com/countries") as url:
    country_data = json.loads(url.read().decode())
    pprint(country_data)
```

```python
covid_data = json_normalize(country_data['data']) 
covid_data.head()
```

```python
covid_data_df = covid_data[['name', 'today.deaths', 'today.confirmed', 'latest_data.confirmed', 'latest_data.deaths', 'latest_data.calculated.death_rate']]
covid_data_df.rename(columns = {'name':'Country', 'today.deaths':'Deaths Today', 'today.confirmed':'Confirmed Today',
                              'latest_data.deaths':'Total Deaths', 'latest_data.confirmed':'Total Confirmed', 'latest_data.calculated.death_rate':'% Death Rate'}, inplace = True)
covid_data_df.set_index(['Country'], inplace = True) 
covid_data_df.round(2)
covid_data_df.head()
```

```python
covid_data_df.to_csv('./Resources/covid_data.csv')
```
