```python
AK_df.plot(kind='line',x='Date',y='death',color='red')
AK_df.plot(kind='line',x='Date',y='positive',color='blue')

plt.show()
```

```python
data1 = AK_df['death']
data2 = AK_df['positive']
d = AK_df['Date']

fig, ax1 = plt.subplots()


ax1.set_xlabel('date')
ax1.set_ylabel('deaths', color='red')
ax1.plot(d, data1, color='red', label='deaths')
ax1.tick_params(axis='y', labelcolor='red')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('positives', color='blue')  # we already handled the x-label with ax1
ax2.plot(d, data2, color='blue', label='positives')
ax2.tick_params(axis='y', labelcolor='blue')

ax1.legend(loc="upper left")
ax2.legend(loc="lower right")


fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
MN_df = state_historical_df.loc[state_historical_df['state'] == 'MN']
MN_df = MN_df[['date', 'positive', 'negative', 'death']]
# AK_df.set_index(['date'], inplace = True) 
MN_df = MN_df.iloc[::-1]
MN_df['Date'] = pd.to_datetime(MN_df['date'].astype(str), format='%Y%m%d')
MN_df
```

```python
d1 = MN_df['death']
d2 = MN_df['positive']
t = MN_df['Date']

plt.plot(t, d1, "-b", label="Deaths")
plt.plot(t, d2, "-r", label="Positive Cases")
plt.legend(loc="upper left")
plt.title('MN Positive Covid-19 Cases and Deaths')
plt.show()
```
